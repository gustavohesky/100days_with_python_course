```python
'''Subscripting'''
print("Hello"[0]) #Adding the index to print a specific letter. It always starts with 0
print("Hello"[-2]) # using negative, it starts the count from the end
```

```python
#String
print("123" + "456") # it will not sum the numbers but it will concatenate them as it's a string

#Integer = Whole number
print(123+456) # It will sum the numbers

#Large numbers

print(123_456_789)

#Float

print(3.14159)

#Boolean
print(True)
print(False)
```

```python
type("Hello") # use type to identify the data type
print(type("Hello"))
print(type(123))
print(type(123.456))
print(type(True))
```

```python
'''Type Conversion'''

int("123") # convert the string to integer
```
