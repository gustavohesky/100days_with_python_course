```python
student_scores ={

'Harry':88,
'Ron':78,
'Hermione':95,
'Draco':75,
'Neville':60
}

student_grades = {}

for each in student_scores:
    if student_scores[each] >= 91:
        student_grades[each] = "Outstanding"
    elif student_scores[each] <= 90 and student_scores[each] >= 81:
        student_grades[each] = "Exceeds Expectations"
    
    elif student_scores[each] <= 80 and student_scores[each] >= 71:
        student_grades[each] = "Acceptable"

    elif student_scores[each] <= 70:
        student_grades[each] = "Fail"

print(student_grades)    

```

```python
# print("\n"*100)

bidders = {}

name = input("What is your name?")

bid = int(input("What is your bid?"))

bidders[name] = bid

more_bids = input("Are there any other bidders (yes/no?) ").lower()

while more_bids == "yes":
    name = input("What is your name?")
    bid = int(input("What is your bid?"))
    bidders[name] = bid

    print("\n"*100)
        
    more_bids = input("Are there any other bidders (yes/no?) ").lower()


def find_highest_bidder(bidders):
    highest_bid = 0
    winner = ""

    for person, amount in bidders.items():
        if amount > highest_bid:
            highest_bid = amount
            winner = person

    print(f"The winner is {winner} with a bid of {highest_bid}")

find_highest_bidder(bidders)
```
