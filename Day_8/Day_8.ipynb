```python
def calculate_love_score(a, b):
    count1 = 0
    count2 = 0

    for each in a:
        if each in ["t", "r", "u", "e"]:
            count1 += 1
        elif each in ["l", "o", "v", "e"]:
            count2 += 1

    for each in b:
        if each in ["t", "r", "u", "e"]:
            count1 += 1
        elif each in ["l", "o", "v", "e"]:
            count2 += 1

    print(str(count1) + str(count2))


calculate_love_score(a="gustavo", b="jessyca")
```

```python
'''Cypher'''


def encrypt(text,shift):
    
    alphabet = [
            "a","b","c","d","e","f","g","h","i","j","k","l","m",
            "n","o","p","q","r","s","t","u","v","w","x","y","z","1","2","3","4","5","6","7","8","9","0"
        ]
    
    result = []
    
    for each in text:
        if each in alphabet:
            index = int(alphabet.index(each))
            new_index = (index + shift) % len(alphabet)
            result.append(alphabet[new_index]) 

    return("".join(result))
    


mode = input("Do you want to encode or decode a message? ").lower()
phrase = input("Type your message: ").lower()
shift = int(input("Please type the shift number: "))

if mode == "decode":
    shift =-shift

result = encrypt(phrase, shift)

print(result)

```
