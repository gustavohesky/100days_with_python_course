```python
## Operators
# == Equal to
# != Not equal to
# > Greater than
# < Less than
# >= Greater than or equal to
# <= Less than or equal to

## Modulo Operator

x = 10 % 5
print(x)
print(f"the remainder is {x}")
x = 10 % 3
print(f"the remainder is {x}")

# Exercise 2 - Even or Odd

number_to_check = int(input("What is the number you want to check? "))
if number_to_check % 2 == 0:
    print(f"{number_to_check} is even")
else:
    print(f"{number_to_check} is odd")
```

```python
# Nested if/ else

# if condition:
#     if another condition:
#         do this
#     else:
#         do this
# else: do this

# Exercise 3

print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))
if height >= 120:
    print("You can ride the rollercoaster")
    age = int(input("What is your age?"))
    if age <= 18:
        print("Please pay €7")
    else:
        print("Please pay €12")
else:
    print("Sorry you have to grow taller before you can ride")
```

```python
# Multiple ifs


print("Welcome to the rollercoaster!")
height = int(input("What is your height in cm? "))
if height >= 120:
    print("You can ride the rollercoaster")
    age = int(input("What is your age?"))
    if age <= 12:
        bill = 5
        print("Child tickets are €5")
    elif age > 12 and age <=18:
        bill = 7
        print("Youth tickets are €7")
    else:
        bill = 12
        print("Adult tickets are €12")
    
    wants_photo = input("Do you want to have a photo? Type y for Yes and n for No")

    if wants_photo == 'y':
        #add €3 to the bill
        bill +=  3
    print(f"Your final bill is €{bill}")

else:
    print("Sorry you have to grow taller before you can ride")
```

```python
# Exercise 5 - Pizza

print("Welcome to Python Pizza Deliveries!")

size = input("What size pizza do you want? S, M or L: ")

price = 0
if size == 'S':
    price += 10
elif size == 'M':
    price += 14
elif size == 'L':
    price += 18
else:
    print("You typed the wrong input")
pepperoni = input("Do you want pepperoni on your pizza? Y or N: ")
if pepperoni == 'Y':
    price += 3
extra_cheese = input("Do you want extra cheese on your pizza? Y or N: ")
if extra_cheese == 'Y':
    price +=2

print(f"Your final bill is €{price}")
```
